#Limpieza de datos

```python
#Queremos saber que columnas tienen valores nulos. Este código recorre las columnas y nos entrega los nombres de las que poseen datos nulos y su cantidad.
a=pokemon.columns.tolist()
for i in pokemon :
  b=pokemon[i]
  if b.isnull().sum():
    print(i,b.isnull().sum())
```

Decisiones para los datos nulos

```python
#Para la columna "height_m", los datos que faltan son reales, por ello se relleno con la mediana. No se utilizo el promedio porque hay outliers que pueden afectar
pokemon["height_m"] = pokemon["height_m"].fillna(pokemon["height_m"].median())
```

```python
#Para la columna "percentage_male", como tiene datos de tipo float, completaremos los nulos con un valor númerico (-1) que representa que el pokemon no tiene genero.
pokemon["percentage_male"] = pokemon["percentage_male"].fillna(-1)
```

```python
#Para la columna "type2" se completo cada dato "NaN" con "no tiene", porque nos entrega información de los pokemones que solo son de 1 tipo y los que poseen segundo tipo
pokemon['type2'] = pokemon['type2'].fillna("no tiene")
```

```python
#Para la columna "weight_kg" ocurre lo mismo que con "height_m", porque faltan datos reales, por ello también se relleno con la mediana
pokemon["weight_kg"] = pokemon["weight_kg"].fillna(pokemon["weight_kg"].median())
```

Estos tipos de gráficos nos permite observar como se distribuyen los valores de attack en cada tipo de poder, donde hay una mayor concentración, ya que son distribuciones. Como queremos calcular el tipo de pokemon con mayor ataque, podemos realizar un promedio de attack por cada tipo de poder 1 y 2.

```python
#Para Type 1
sns.barplot(data=pokemon, x="type1", y="attack", estimator="mean",
            order=pokemon.groupby("type1")["attack"].mean().sort_values(ascending=False).index)

plt.xticks(rotation=45)
plt.title("Ataque promedio para los type1")
plt.show()
```

```python
#Para Type 2, exceptuando los que no tienen segundo poder
sns.barplot(data=pokemon[pokemon["type2"]!="no tiene"], x="type2", y="attack", estimator="mean",
            order=pokemon[pokemon["type2"]!="no tiene"].groupby("type2")["attack"].mean().sort_values(ascending=False).index)

plt.xticks(rotation=45)
plt.title("Ataque promedio para los type2")
plt.show()
```

##Correlación entre las variables numéricas

```python
df = pokemon

# Calcular correlación (solo números)
corr = df.corr(numeric_only=True)    ## uso de IA

# Dibujar el heatmap
plt.figure(figsize=(18,16))      #tamaño de la figura

# se pinta la matriz en un mapa de colores
# se fija un rango de colores
# define el esquema de colores
# definir número de decimales
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1, fmt=".1f") ## uso de IA
plt.title("Heatmap")
plt.show()


## IA ## se uso la IA para la explicacion funciomaniento sns.heatmap y ver sus caracteristicas
```

# Análisis De Correlación

```python
p = pokemon
# Calcular correlación (solo números)
corr = p.corr(numeric_only=True)
print(corr)
```

Para la interpretación de valores de la correlacion, sabemos que se aplica cierto criterio para evaluar su relacion.

* Una correlación cercana a 0: indica que no hay una relación lineal
aparente entre las dos variables.
* Una correlación entre 0.3 y 0.5: se considera moderada.
* Una correlación entre 0.5 y 0.7: se considera fuerte.
* Una correlación mayor a 0.7: se considera muy fuerte.
* Todos los valores en diagonal con 1 son la línea de las mismas variables en vertical y horizontal y no deben tenerse en cuenta.

Esto evitara perdida de tiempo y gasto innecesarios(en caso de que sea)

considerar que Is_legendary y Base_eggs_steps tienen la mayor correlacion, sin embargo, en el mundo de pokemon esto no tienen sentido, ya que, los pokemones lengendarios no colocan huevos, por lo que la correlacion no indica causalidad












Se pude apreciar que la Base_total se relaciona con varios de nuestros parametros.
En pokemon la base_total corresponde a la suma de 6 estadistica base (ataque,defensa,velocidad, etc)
